```python
from nltk.stem import SnowballStemmer

# Initialize SnowballStemmer for English
stemmer = SnowballStemmer("english")

words = ["running", "flies", "happiness", "caring", "smiling", "jumps", "quickly"]

# Apply Snowball stemming
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

# lemmatization


```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

words = ["running", "flies", "happiness", "caring", "better", "ate"]

# Apply lemmatization
lemmatized_words = [lemmatizer.lemmatize(word, pos="v") for word in words]  # 'v' for verb , n for noun , adjectve-a, adverb -r

print(lemmatized_words)
```

# stopwords


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')

# Load English stopwords
stop_words = set(stopwords.words("english"))

# Example sentence
text = "This is an example sentence, and it includes some stopwords."

# Tokenize words
words = word_tokenize(text)

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Download necessary datasets
nltk.download('stopwords')
nltk.download('punkt')

# Initialize Stopwords & Stemmer
stop_words = set(stopwords.words("english")) ##things tht have to remove from pargraph
stemmer = PorterStemmer()

# Example sentence
text = "Running quickly in the park is enjoyable, but the weather is changing."

# Tokenize words
words = word_tokenize(text)
print(words)

# Remove stopwords and apply stemming
filtered_stemmed_words = [stemmer.stem(word) for word in words if word.lower() not in stop_words]

print(filtered_stemmed_words)

```

```python
#  In this **Continuous Bag of Words (CBOW) model**, we are trying to predict the missing word (e.g., "is") based on surrounding words (context).

# ### **How CBOW Works Here:**
# 1. **Input:**
#    Given the sentence:
#    **"ineuron company is related to data science"**,
#    if we remove the word **"is"**, the model takes the surrounding words:
#    **["ineuron", "company", "related", "to"]** as input.

# 2. **Hidden Layer Processing:**
#    - These words are converted into **one-hot vectors** (represented as 7×5 matrices in the diagram).
#    - Weighted sum is computed and passed through the hidden layer.
#    - The model learns patterns by adjusting **weights** during training.

# 3. **Output Layer:**
#    - The model predicts the **most probable missing word** (e.g., "is").
#    - Softmax activation function assigns probabilities to all words, and the word with the highest probability is chosen.

# ### **Goal of This Model:**
# Yes! We are **learning how words relate to each other** based on context.
# This helps in **word embeddings** like Word2Vec, where similar words have closer numerical representations in vector space.

```
